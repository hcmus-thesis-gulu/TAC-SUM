```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
context_path = "/content/drive/MyDrive/HCMUS/Y4/Thesis/Documents/Report/Thesis/dino-b16-cls-euclidean-pca-34-tsne-2/contexts"
summary_path = "/content/drive/MyDrive/HCMUS/Y4/Thesis/Documents/Report/Thesis/dino-b16-cls-euclidean-pca-34-tsne-2/summaries"
embedding_path = "/content/drive/MyDrive/HCMUS/Y4/Thesis/Documents/Report/Thesis/dino-b16-cls-euclidean-pca-34-tsne-2/embeddings"
video_path = "/content/data/videos"
```

```python
# filename = "Fire Domino"
filename = "Jumps"
# filename = "Paintball"
```

```python
context_file = context_path + f"/{filename}_segments.npy"
kf_file = summary_path + f"/{filename}_keyframes.npy"
score_file = summary_path + f"/{filename}_scores.npy"
sample_file = embedding_path + f"/{filename}_samples.npy"
video_file = video_path + f"/{filename}.webm"
```

```python
keyframes = np.load(kf_file)
print(keyframes)
scores = np.load(score_file)
print(scores)
```

```python
samples = np.load(sample_file)
print(samples)
```

```python
segments = np.load(context_file)
print(segments.shape)
```

```python
sampled_scores = []

for _, start, end in segments:
    pick = samples[(start + end) // 2]
    score = end - start + 1
    sampled_scores.append((pick, score))
```

```python
print(sampled_scores)
```

```python
thresh = np.array(sampled_scores).mean(axis=0)[-1]
print(thresh)
```

```python
sorted_scores = sorted(sampled_scores, key=lambda tup: -tup[-1])
print(sorted_scores)
filtered_scores = [sorted_score for sorted_score in sorted_scores if sorted_score[-1] > thresh]
print(filtered_scores)
```

```python
sampled_indexes = []

target_indexes = [filtered_score[0] for filtered_score in filtered_scores]
sampled_indexes = sorted(target_indexes)
print(sampled_indexes)
```

```python
interval = 5

# Set the maximum number of frames per row
r = 8
```

```python
# Load the video file
video = cv2.VideoCapture(video_file)

# Calculate the frame rate of the video
fps = video.get(cv2.CAP_PROP_FPS)

# Calculate the frame interval for the first set of frames
frame_interval = int(fps * interval)

# Calculate the total number of frames in the video
frame_count = int(video.get(cv2.CAP_PROP_FRAME_COUNT))

# Generate the first set of frame indexes
original_indexes = np.arange(0, frame_count, frame_interval)

# Calculate the number of rows for each set of frames
rows1 = (len(original_indexes) + r - 1) // r
rows2 = (len(sampled_indexes) + r - 1) // r

# Set the number of rows and columns for the grid
rows = rows1 + rows2
cols = r

# Set the size of the figure
figsize = (cols * 4, rows * 4)

# Create a figure and axes for the plot
fig, axes = plt.subplots(figsize=figsize, nrows=rows, ncols=cols)

# Add a title to the center top of the plot
fig.suptitle(f"Comparison between {interval}-second representatives and summarized keyframes of an input",
             fontsize=24)

# Adjust the subplot parameters to reduce space between rows
plt.subplots_adjust(hspace=0.1)

# Loop through the first set of frame indexes
for i, frame_index in enumerate(original_indexes):
    # Set the current position in the video
    video.set(cv2.CAP_PROP_POS_FRAMES, frame_index)

    # Read the current frame from the video
    ret, frame = video.read()

    # Convert the frame from BGR to RGB
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Calculate the row and column for the current frame
    row = i // cols
    col = i % cols

    print(f"Frame {i + 1}@{frame_index} positioned at x={row}, y={col}")

    # Plot the current frame on the grid
    axes[row, col].imshow(frame)
    axes[row, col].set_title(f"{i + 1}. Frame {frame_index}")
    axes[row, col].axis('off')

print("Completed positioning original video!")

# Loop through the second set of frame indexes
for i, frame_index in enumerate(sampled_indexes):
    # Set the current position in the video
    video.set(cv2.CAP_PROP_POS_FRAMES, frame_index)

    # Read the current frame from the video
    ret, frame = video.read()

    # Convert the frame from BGR to RGB
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Calculate the row and column for the current frame
    row = rows1 + i // cols
    col = i % cols

    print(f"Frame {i + 1}@{frame_index} positioned at x={row}, y={col}")

    # Plot the current frame on the grid
    axes[row, col].imshow(frame)
    axes[row, col].set_title(f"{i + 1}. Keyframe @ {frame_index}")
    axes[row, col].axis('off')

print("Completed positioning sampled video!")

# Center-align any incomplete rows by hiding unused axes and shifting remaining axes to center position.
for i in range(rows):
    count_data_axes_in_row_i = sum([axes[i,j].has_data() for j in range(cols)])
    print(f"Row {i} has {count_data_axes_in_row_i} frames")

    if count_data_axes_in_row_i < cols:
        shift_axes_by_i = (cols - count_data_axes_in_row_i) // 2
        print(f"Row {i} shifts {shift_axes_by_i} to the right")

        for j in range(cols-1,-1,-1):
            if axes[i,j].has_data():
                print(f"Row {i} moves {j} to {j+shift_axes_by_i}")
                if shift_axes_by_i > 0:
                    # Move the image and title to the new position
                    axes[i,j+shift_axes_by_i].imshow(axes[i,j].images[0].get_array())
                    axes[i,j+shift_axes_by_i].set_title(axes[i,j].get_title())
                    axes[i,j+shift_axes_by_i].axis('off')

                    # Clear the old position
                    axes[i,j].clear()
                    axes[i,j].axis('off')
            else:
                axes[i,j].axis('off')

print("Completed centering short rows!")

# # Add a horizontal line separating the two sets of frames
# plt.axhline(y=rows1 - 0.5, color='black', linestyle='--')

# Add a legend explaining the horizontal line
plt.figtext(0.5, 0.5,
            'Above frames are original representatives and below frames are summarized keyframes',
            fontsize=16, ma='center', ha='center', va='center')

# # Show the plot
# plt.show()

# Release the video capture object
video.release()

plt.savefig("/content/drive/MyDrive/HCMUS/Y4/Thesis/Documents/Report/report/content/experiment/figures/video-visual.svg",
            format='svg', bbox_inches='tight')
plt.savefig("/content/drive/MyDrive/HCMUS/Y4/Thesis/Documents/Report/report/content/experiment/figures/video-visual.png",
            format='png', bbox_inches='tight')
```
